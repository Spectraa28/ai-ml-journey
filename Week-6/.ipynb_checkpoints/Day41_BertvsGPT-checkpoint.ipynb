```python
from transformers import pipeline

# BERT for fill-mask (its actual pretraining task)
bert_fill = pipeline('fill-mask', model='bert-base-uncased')

# GPT-2 for text generation
gpt2_gen = pipeline('text-generation', model='gpt2')

# BERT predicts masked token
bert_result = bert_fill("The animal didn't cross the street because [MASK] was too tired.")
print("BERT fill-mask predictions:")
for r in bert_result[:3]:
    print(f"  {r['token_str']}: {r['score']:.4f}")

# GPT-2 generates continuation
gpt2_result = gpt2_gen("The animal didn't cross the street because", 
                        max_new_tokens=20, 
                        num_return_sequences=1)
print("\nGPT-2 continuation:")
print(gpt2_result[0]['generated_text'])
```

## Summary

**What I built:** Hands-on comparison of BERT and GPT-2 using 
HuggingFace pipelines — demonstrated understanding vs generation.

**Key results:**
- BERT fill-mask: predicted "it" with 75.56% confidence (correct)
- GPT-2 generation: grammatically correct but semantically wandering

**BERT vs GPT architecture:**
- BERT: encoder, bidirectional attention, sees full context both ways
  - Pretrained with Masked Language Modeling
  - Used for: classification, embeddings, Q&A, NER
  - Can't generate: trained on full context, not next-token prediction
- GPT: decoder, causal/masked attention, sees only previous tokens
  - Pretrained with next-token prediction
  - Used for: text generation, code completion, chat
  - Can't fill-mask: never trained to use future context

**Why GPT-2 generation is incoherent:**
- Small model (117M parameters)
- No instruction fine-tuning or RLHF
- Raw next-token prediction without alignment
- Modern GPT-4/Claude add RLHF on top

**Key insight:** Architecture follows task. Understanding needs 
full context (bidirectional). Generation needs left-only context 
(causal) to avoid cheating during training. You can't swap them.
