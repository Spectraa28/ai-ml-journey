```python
import numpy as np 
```

```python
# MInimal Self attention from Scratch in Numpy 
```

# Self attention 

```python
np.random.seed(42)
X = np.random.randn(4,8) # 4 words , embedding dim =  8 
```

```python
# Weight metrices 
W_Q   =  np.random.randn(8,8)
W_K   =  np.random.randn(8,8)
W_V   =  np.random.randn(8,8)
```

```python
#calculating q , k , v (query ,  key , value )
Q = X @ W_Q
K = X @ W_V
V = X @ W_V
```

```python
print(f"Q shape: {Q.shape}")
print(f"K shape: {K.shape}")
print(f"V shape: {V.shape}")
```

```python
# Computing attention scores 

# Step 1 compute raw attention scores - each words query dots with every other words key 
# Q @ k.T gives a 4,4 matrix - evey word vs every word 
# scores[i][j] = how much should word i attend to word j

scores  = Q @ K.T
```

```python
# Step 2 : scaled the scores 
# Large values PuSh softmax into regions with tiNy gradientS (vanishing gradients only)
# Scaling by sqrt(d_k) keeps values in a stabble Range 

d_k = Q.shape[1]
scores =  scores/ np.sqrt(d_k) # scaled scores
```

```python
# Step 3  Softmax - convert scores to probabilities 
def softmax(x):
    e_x = np.exp(x - np.max(x, axis = -1 , keepdims=True))
    return e_x / e_x.sum(axis=-1 , keepdims=True)

attention_weights  = softmax(scores)


    
# Step 4 - Wieghted sum  of values 
output = attention_weights @ V
```

```python
print("Attention weights (each row sums to 1):")
print(np.round(attention_weights, 3))
print(f"\nOutput shape: {output.shape}")
```
