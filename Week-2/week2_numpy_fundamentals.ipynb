```python
# Python list - each element is a Python object
# stored randomly in memory
python_list = [1, 2, 3, 4, 5]

# NumPy array - all elements same type
# stored in continuous memory block
numpy_array = np.array([1, 2, 3, 4, 5])

print("Python list type:", type(python_list[0]))
print("NumPy array dtype:", numpy_array.dtype)
print("NumPy array shape:", numpy_array.shape)
print("NumPy array size:", numpy_array.size)
print("NumPy memory (bytes):", numpy_array.nbytes)
```

Python wastes memory storing type information for every single element. NumPy stores type once for the entire array.
That's why NumPy is faster — less memory = faster access = faster computation.

# ML connection

```python
# This is what happen in Logistic Regression

# Fake wieght
weight = np.array([0.5,-2.1,0.3])
```

```python

passenger = np.array([2,0,25]) #Pclass, Sex,  Age
```

```python
z = np.dot(weight,passenger)
print("Weighted sum z:", z)
```

```python
# Sigmoid function = converts  to probability 

def signoid(x):
    return 1 / (1+ np.exp(-x))
```

```python
probability = signoid(z)
```

```python

print("Survival Probability : ", round(probability * 100, 1), "%")
```

# By model

```python
weights = np.array([-0.5528, 2.1375, -0.0313, 
                    -0.6255, -0.1250, 0.0160, 0.7433])
```

```python
# Sonu's features
sonu = np.array([2,0,21,2,1,20,0])
```

```python
# Dot product 
z = np.dot(weights,sonu) + 0.5788779288662111
probability = signoid(z)
print("Sonu Survival probability: ", round(probability * 100 , 1), "%")
```

```python
# Her features
her = np.array([1,1,20,0,0,100,1])
```

```python

z = np.dot(weights,her) + 0.5788779288662111
```

```python
probability = signoid(z)
```

```python
print("Shree Survival Probability:", round(probability * 100,1), "%")
```

```python
# Logistic Regression is just:
# 1. Weighted sum of features + intercept
# 2. Sigmoid squashes result to 0-1
# 3. Threshold at 0.5 for binary prediction
#
# model.coef_     = learned weights
# model.intercept_ = learned bias
# np.dot()        = weighted sum
# sigmoid()       = probability conversion
#
# Everything else is just finding the right weights
# That process = gradient descent (Week 3)
```
