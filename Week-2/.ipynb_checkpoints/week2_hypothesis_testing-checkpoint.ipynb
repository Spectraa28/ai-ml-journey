# Setup

```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
```

# Loading and cleaning the data

```python
tc = pd.read_csv("./titanic/train.csv")
```

```python
tc['Age'] = tc['Age'].fillna(tc['Age'].median())
tc['Has_cabin'] = tc['Cabin'].apply(lambda x: 0 if pd.isnull(x) else 1)
tc = tc.drop('Cabin',axis=1)
tc['Sex'] = tc['Sex'].map({'female':1,'male':0})
```

```python
survived_ages = tc[tc['Survived'] == 1]['Age']
died_ages = tc[tc['Survived'] == 0]['Age']
```

```python
print("Survivors - Mean age:", round(survived_ages.mean(), 2))
print("Died - Mean age:", round(died_ages.mean(), 2))
print("Difference:", round(survived_ages.mean() - died_ages.mean(), 2))
```

# t-test

```python
t_stat , p_value = stats.ttest_ind(survived_ages,died_ages)
```

```python
print("T-statistic: ", round(t_stat,4))
print("p-value", round(p_value,4))
```

```python
if p_value < 0.05:
    print("Result: SIGNIFICANT difference")
    print("The age difference is probably real, not random chance")
else:
    print("Result: NOT significant")
    print("The age difference could be random chance")
```

```python
female_survival = tc[tc['Sex'] == 1]['Survived']
male_survival = tc[tc['Sex'] == 0]['Survived']
```

```python
t_stat, p_value = stats.ttest_ind(female_survival, 
                                   male_survival)
```

```python
print("Female survival rate:", round(female_survival.mean(), 3))
print("Male survival rate:", round(male_survival.mean(), 3))
print("T-statistic:", round(t_stat, 4))
print("P-value:", p_value)

if p_value < 0.05:
    print("\nResult: SIGNIFICANT")
else:
    print("\nResult: NOT significant")
```

```python
cabin_survival = tc[tc['Has_cabin'] == 1]['Survived']
no_cabin_survival = tc[tc['Has_cabin'] == 0]['Survived']
```

```python
t_stat, p_value = stats.ttest_ind(cabin_survival, 
                                   no_cabin_survival)
```

```python
print("Has cabin survival rate:", 
      round(cabin_survival.mean(), 3))
print("No cabin survival rate:", 
      round(no_cabin_survival.mean(), 3))
print("P-value:", p_value)
```

```python
if p_value < 0.05:
    print("Result: SIGNIFICANT")
else:
    print("Result: NOT significant")
    
```

```python
# Hypothesis testing summary:
# Sex vs Survival:       p = 1.4e-69 → significant
# Has_Cabin vs Survival: p = 3e-22   → significant  
# Age vs Survival:       p = 0.053   → NOT significant
#
# Key insight: statistical significance predicted
# model feature weights before training
# Statistics and ML tell the same story
# Day 11: Hypothesis testing - p-values validate feature engineering decisions
```
