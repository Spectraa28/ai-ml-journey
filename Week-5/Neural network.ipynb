```python
import numpy as np 
```

# XOR Dataset 

```python
X = np.array([[0,0],
             [0,1],
              [1,0],
              [1,1]
             ])
```

```python
y = np.array([
    [0],[1],[1],[0]
])
```

# Forward Pass 

```python
def sigmoid(z):
    s = 1/(1+ np.exp(-z))
    return s
```

# INitializing

```python
np.random.seed(42)

W1 = np.random.randn(2,4)
b1 = np.zeros((1,4))      
W2 = np.random.randn(4,1)
b2 = np.zeros((1,1))       
```

```python
# Forward pass function 
```

```python
def forward(X,W1,b1,W2,b2):
    # LAyer 1 : Input to hiden 
    z1 = X @ W1 + b1
    a1 = sigmoid(z1)

    # LAyer 2 : hiden to Output 
    z2 = a1 @ W2 + b2
    a2 = sigmoid(z2)

    return a1, a2
```

```python
a1, a2 = forward(X, W1, b1, W2, b2)
print("Predictions before training:")
print(a2)
```

```python
def compute_loss(y_true, y_pred):
    m = y_true.shape[0]  
    loss = -(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))
    return loss.mean()
```

```python
loss = compute_loss(y, a2)
print(f"Loss before training: {loss}")
```

```python
def backward(X, y, a1, a2, W2):
    m = X.shape[0]
    
    # Output layer gradients
    dz2 = a2 - y
    dW2 = a1.T @ dz2 / m
    db2 = dz2.sum(axis=0, keepdims=True) / m
    
    # Hidden layer gradients
    dz1 = (dz2 @ W2.T) * a1 * (1 - a1)
    dW1 = X.T @ dz1 / m
    db1 = dz1.sum(axis=0, keepdims=True) / m
    
    return dW1, db1, dW2, db2
```

```python
dW1, db1, dW2, db2 = backward(X, y, a1, a2, W2)
print("Gradient shapes:")
print(f"dW1: {dW1.shape}, db1: {db1.shape}")
print(f"dW2: {dW2.shape}, db2: {db2.shape}")
```

```python
learning_rate = 1.0
epochs = 10000
```

```python
losses = []
```

```python
for epoch in range(epochs):
    # Forward pass
    a1, a2 = forward(X, W1, b1, W2, b2)
    
    # Compute loss
    loss = compute_loss(y, a2)
    losses.append(loss)
    
    # Backward pass
    dW1, db1, dW2, db2 = backward(X, y, a1, a2, W2)
    
    # Update weights
    W1 -= learning_rate * dW1
    b1 -= learning_rate * db1
    W2 -= learning_rate * dW2
    b2 -= learning_rate * db2
    
    # Print progress
    if epoch % 1000 == 0:
        print(f"Epoch {epoch}, Loss: {loss:.4f}")

print("\nFinal predictions:")
a1, a2 = forward(X, W1, b1, W2, b2)
print(a2)
print("\nTarget:")
print(y)
```
