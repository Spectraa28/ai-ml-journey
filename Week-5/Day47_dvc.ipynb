## Summary

**What I built:** DVC data versioning — track datasets like Git 
tracks code, restore any previous version instantly.

**Key concepts:**
- DVC stores actual data in remote storage (S3, GCS, local)
- Git stores only the .dvc pointer file (tiny, contains MD5 hash)
- MD5 hash fingerprints exact file contents — one row change = new hash
- dvc add → creates pointer file, moves data to DVC cache
- dvc push → uploads data to remote storage
- dvc checkout → restores exact dataset for current Git commit

**The workflow:**
1. dvc add dataset.csv → creates dataset.csv.dvc
2. git commit dataset.csv.dvc → versions the pointer
3. Data changes → dvc add again → new hash → new git commit
4. git checkout old_commit dataset.csv.dvc → dvc checkout → original data restored

**Why this matters:**
- Git alone: can't version large files, breaks on CSVs > 100MB
- DVC: handles any size, stores in cloud, restores in seconds
- Every experiment is reproducible — same code + same data = same model
