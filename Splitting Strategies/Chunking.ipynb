````python
 import re
````

````python
def manual_text_splitter(text,chunk_size=100,chunk_overlap=0):
    # Simplified patterns to avoid "multiple repeat" errors
    separators = [
        r"\n#{1,6} ",         # Headers (added a space for standard MD)
        r"```\n",             # Code blocks
        r"\n\*{3,}\n",        # Horizontal Lines (***)
        r"\n-{3,}\n",         # Horizontal Lines (---)
        r"\n_{3,}\n",         # Horizontal Lines (___)
        r"\n\n",              # Double new lines
        r"\n",                # New line
        r" "                  # Spaces
    ]
    
    combined_pattern = f"({'|'.join(separators)})"
    
    parts = re.split(combined_pattern,text)
    parts = [p for p in parts if p]
    
    chunks = []
    current_chunk = ""
    
    for part in parts:
        if len(current_chunk) + len(part) <= chunk_size:
            current_chunk += part 
        else:
            if current_chunk:
                chunks.append(current_chunk.strip())
                
                # Start new chunk 
            if chunk_overlap > 0 and len(current_chunk) > chunk_overlap:
                current_chunk = current_chunk[-chunk_overlap:] + part
            else:
                current_chunk  = part
                
    if current_chunk:
        chunks.append(current_chunk.strip())
        
    return chunks
````

````python
markdown_text = """# Introduction
This is a test.
## Subheader
Here is some more text that we want to split.
"""
````

# Pdf format documents  with table 
 WE will use a new library to handle it because in pure py its just to complex to handle those that library is either unstructured or docling


````python
import os
from pathlib import Path
from docling.document_converter import DocumentConverter, PdfFormatOption
from docling.datamodel.pipeline_options import PdfPipelineOptions
from docling.datamodel.base_models import InputFormat
from docling.chunking import HybridChunker
from docling.datamodel.pipeline_options import AcceleratorOptions, AcceleratorDevice
````

````python
def process_universal_pdf(pdf_path, output_dir="output"):
    output_path = Path(output_dir)
    output_path.mkdir(exist_ok=True)
    images_path = output_path / "images"
    images_path.mkdir(exist_ok=True)

    pipeline_options = PdfPipelineOptions()
    pipeline_options.generate_picture_images = True  # Capture images/charts
    pipeline_options.do_ocr = True                   # Fallback for scanned PDFs
    
    pipeline_options.accelerator_options = AcceleratorOptions(
    device=AcceleratorDevice.CPU, 
    num_threads=4
    )
    converter = DocumentConverter(
        format_options={
            InputFormat.PDF: PdfFormatOption(pipeline_options=pipeline_options)
        }
    )

    # 2. Convert the Document
    result = converter.convert(pdf_path)
    doc = result.document

    # 3. Save Images/Figures
    for i, element in enumerate(doc.pictures):
        image = element.get_image(doc)
        image.save(images_path / f"figure_{i}.png")

    # 4. Extract Markdown (with Table structure)
    full_markdown = doc.export_to_markdown()
    with open(output_path / "document.md", "w") as f:
        f.write(full_markdown)

    
    chunker = HybridChunker()
    chunk_list = list(chunker.chunk(doc))

    return {
        "markdown": full_markdown,
        "chunks": [c.text for c in chunk_list],
        "tables_found": len(doc.tables),
        "images_saved": len(doc.pictures)
    }
````

````python
results = process_universal_pdf("2025 Q3 Financial Statements (consolidated).pdf")
````

````python
# Print the first 3 chunks to see what they look like
for i, chunk in enumerate(results["chunks"][:3]):
    print(f"--- Chunk {i} ---")
    # If your function returned 'c.text', use this:
    print(f"Content: {chunk[:1000]}...") 
    print("-" * 30)

# Check total number of chunks
print(f"Total chunks created: {len(results['chunks'])}")
````

# Semantic chunking 

````python
import numpy as np 
from sentence_transformers import SentenceTransformer
import re
````

````python
def semantic_chunker(text,threshold_percentile=95):
    model  = SentenceTransformer('all-MiniLM-L6-v2')
    
    # Split into sentences 
    sentences  = re.split(r'(?<=[.!?]) +',text)
    if len(sentences) < 2:
        return sentences
    
    # GEt embeddings 
    embeddings = model.encode(sentences)
    
    # COsine distances between consecutive sentences 
    distances = []
    for i in range(len(embeddings) - 1):
        # COsine similarity 
        sim = np.dot(embeddings[i], embeddings[i+1])/ (np.linalg.norm(embeddings[i]) * np.linalg.norm(embeddings[i+1]))
        distances.append(1-sim) # Distance is 1 - similarity 
        
    # Determine Breakpoints 
    breakpoint_threshold = np.percentile(distances, threshold_percentile)
    indices_above_truth = [i for i, x in enumerate(distances) if x >breakpoint_threshold]
    
    chunks = []
    start_index = 0
    for index in indices_above_truth:
        chunk = " ".join(sentences[start_index:index+1])
        chunks.append(chunk)
        start_index = index + 1
        
    chunks.append(" ".join(sentences[start_index:]))
    
    return chunks
````

````python
text = "The economy is showing signs of recovery. Interest rates are expected to stay flat. On a different note, the new space telescope launched yesterday. It will capture images of distant galaxies."
my_chunks = semantic_chunker(text, threshold_percentile=80)

for i, chunk in enumerate(my_chunks):
    print(f"Chunk {i}: {chunk}\n")
````
