# data Setup

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
```

```python
x = np.linspace(0,1,20)
y = 3 * x + np.random.normal(0,0.3,20)
```

```python
print("X shape:", x.shape)
print("y shape:", y.shape)
```

```python
weight = 0.0
learning_rate = 0.01
alpha = 0.1
epochs = 100
```

```python
loss_history = []
weight_history = []
```

```python
for epoch in range(epochs):
    predictions = weight * x
    error = predictions - y

    # Ridge loss (MSE + PENALTY)
    ridge_loss =  np.mean(error**2) + alpha * (weight**2)
    loss_history.append(ridge_loss)

    # ridge gradient
    gradient = (2/len(x)) * np.sum(error * x) + (2 * alpha * weight)

    weight = weight - learning_rate * gradient
    weight_history.append(weight)

    if epoch % 10 == 0:
        print(f"Epoch {epoch:3d} |  Weight: {weight:.4f} | Loss : {ridge_loss:.4f} ")

print(f"\nFinal weight: {weight:.4f}")
print(f"True weight: 3.000")
```

```python
alphas = [0.1, 0.01, 0.001, 0.0001]

for a in alphas:
    weight = 0.0
    for epoch in range(100):
        predictions = weight * x
        errors = predictions - y
        gradient = (2/len(x)) * np.sum(errors * x) + (2 * a * weight)
        weight = weight - learning_rate * gradient
    print(f"Alpha: {a:<8} Final weight: {weight:.4f}")

print(f"\nTrue weight: 3.0000")
```

```python
learning_rates = [0.01, 0.1, 0.5, 1.0]
alpha = 0.0001

for lr in learning_rates:
    weight = 0.0
    for epoch in range(100):
        predictions = weight * x
        errors = predictions - y
        gradient = (2/len(x)) * np.sum(errors * x) + (2 * alpha * weight)
        weight = weight - lr * gradient
    print(f"LR: {lr:<6} Final weight: {weight:.4f}")

print(f"\nTrue weight: 3.0000")
```
